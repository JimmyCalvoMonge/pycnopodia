```python
from pycno2 import PycnoSIX2
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_path = './EnsoData/ENSO_sim_1.csv'
data = pd.read_csv(data_path)
plt.plot(data['x'])
plt.show()
```

```python
pycnoObj = PycnoSIX2(
    x00=[0.2,0.3,0.1,0.4],
    mu_I=0.7, mu_S=0.6,
    delta=0.65, k=1,
    lambda_J=0.34, lambda_A=0.76,
    r_b=0.5, gamma_r=0.1,
    alpha_b=0.5, gamma_alpha=0.1,
    beta_b=0.5, gamma_beta=0.1,
    ensi_file_path=data_path
)
pycnoObj.solve_odes_system_odeint()
pycnoObj.plot_ode_solution()
pycnoObj.plot_dN_dt()
```
